```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr
import numpy as np
```

```python
# File Path
BASE_DIR = Path().resolve().parents[1]
file_PATH = BASE_DIR / "data" / "processed" / "final_arctic.csv"

# Load the merged dataset
df = pd.read_csv(file_PATH)
```

```python
cols_of_interest = [
    "thick_m", "cdir", "uvb",	"slhf",	
    "sf", "t2m", "sst","istl1", "istl2", "istl3",	"istl4", "skt",	"ishf", 
    "tcc", "tsn"

]

df_corr_data = df[cols_of_interest].dropna()

print("‚úÖ Correlation-ready rows:", len(df_corr_data))
```

```python
# 1Ô∏è‚É£ Pearson Correlation Matrix
pearson_corr = df_corr_data.corr(method="pearson")

# üí† Extract correlations of interest

```

```python
# Heatmap of Pearson Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Pearson Correlation Heatmap (World)")
plt.tight_layout()
plt.show()
```
